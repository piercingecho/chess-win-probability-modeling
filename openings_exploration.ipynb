```python
import pandas as pd
```

```python
df = pd.read_csv("base_chess_data.csv")
df = df.drop(["Unnamed: 0"], axis=1)
df.head()
```

```python
# add first move as a column to each game

opening_moves = []

for index, row in df.iterrows():
    opening_moves.append(row['Moves'].split(" ")[0])


df['OpeningMove'] = opening_moves
df.head()
```

```python
# add whether white and black castle to each game

white_castles = []
black_castles = []

for index, row in df.iterrows():
    white_castles.append(("O-O" in row['Moves'].split(" ")[::2]))
    black_castles.append(("O-O" in row['Moves'].split(" ")[1::2]))



df['WhiteCastles'] = white_castles
df['BlackCastles'] = black_castles
df.head()
```
